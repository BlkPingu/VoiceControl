```python
out_list = []

metadata = pd.read_csv('metadata_survey.csv')

def get_metadata(directory):
    for i, row in metadata.iterrows():
        if row['uuid'] in directory:
            return row['uuid']

for wav_path in wav_paths:
    for i, row in questions.iterrows():
        if row['ID'] == wav_path[114+83:-4]:
            res = ({'ID':row['ID'], 
                    'EXPECTED_RESULT':row['RESULT'],
                    'EXPECTED_TEXT':row['TEXT'],
                    'IDEA':row['IDEA'],
                    'TYPE':row['TYPE'],
                    'PATH': wav_path,
                    'UUID': get_metadata(wav_path)
                   })
            out_list.append(res)
```

```python
out_df = pd.DataFrame(out_list)
```
